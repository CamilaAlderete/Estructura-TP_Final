```ruby
def limpia_texto(df)
  
  poco_significativas = ['como','pero','ya','todo','todos','todas','solo','esto','entre','otro','en','a','e','i','o','u','ni','ya','ese','eso','esos','tan','sin','no','sí','si','al','para','el', 'ella', 'ellos', 'ellas', 'él', 'la', 'las','lo','los', 'le','les', 'y', 'o', 'a', 'me', 'mí', 'mi', 'tu','tus','tú', 'te','ti','se', 'de','del','su','sus', 'mas','más', 'un','unos', 'este', 'en', 'una', 'por', 'que','qué','es', 'rt','con','vs','tw']

  patron_poco_sig = /\b(como|pero|ya|todo|todos|todas|solo|esto|entre|otro|en|a|e|i|o|u|ni|ya|ese|eso|esos|tan|sin|no|sí|si|al|para|el|ella|ellos|ellas|él|la|las|lo|los|le|les|yo|ya|me|mí|mi|tu|tus|tú|te|ti|se|de|del|su|sus|mas|más|un|unos|este|y|en|una|por|que|qué|es|rt|con|vs|tw)\b/


    for i in(0...df.size )
         texto = df.row[i][0]

         #emojis y otras weas
         texto.gsub!(/[\u{10000}-\u{10FFFF}]/, '')
         texto.gsub!(/[\u{1f300}-\u{1f5ff}]/, '')
         texto.gsub!(/[\u{2500}-\u{2BEF}]/, '')
         texto.gsub!(/[\u{1f600}-\u{1f64f}]/, '')
         #texto.gsub!(/[^\u{1F600}-\u{1F6FF}\s]/, '')

        #menciones y hashtags
        texto.gsub!(/\B[@#]\S+\b/, '')

        #urls
        texto.gsub!(/http.+/, '')
        
        #este no.....
        #texto.gsub!(/http.+ /, '')
      
        #espacios
        texto.gsub!(/\s+/, ' ')

        #digitos
        texto.gsub!(/[[:digit:]]/, '')

        #puntuaciones
        texto.gsub!(/[[:punct:]]/, ' ')

        texto.gsub!(patron_poco_sig,' ')
        #poco_significativas.each do |palabra|
        #  texto.gsub!(/ #{palabra} /,' ')
        #end

        #texto.gsub!(/\\b\\w{1,3}\\s/, ' ') #no anda....

        df.row[i][0] = texto

    end

    return df
  
end
```

```ruby
def arma_conjunto(df)
    set = Array.new
  
    #descarta palabras muy cortas
    for i in(0...df.size )
      texto = df.row[i][0].split

      for j in(0...texto.size )
          if texto[j].length > 3
              set.push(texto[j])
          end  
      end
    end    

    return set
end
```

```ruby
#https://nbviewer.org/github/SciRuby/sciruby-notebooks/blob/master/Data%20Analysis/Usage%20of%20Vector.ipynb

def top_30(set_palabras, set_frec)
    set_palabra_top_30 = Array.new
    set_frec_top_30 = Array.new

    #top 30
    for j in(0...15 )
         set_palabra_top_30.push(set_palabras[j])
         set_frec_top_30.push(set_frec[j])      
    end


    #Daru.plotting_library = :gruff
    v = Daru::Vector.new(set_frec_top_30, index:set_palabra_top_30, name: "")
    return v.plot(type: :bar) do |plt|
      plt.x_label "Palabra"
      plt.y_label "Numero"

    end

end
```

## Número de personas diferentes (usuarios) que escribieron tweets por fecha.

```ruby
usuarios_y_fechas = Array.new

#fechas y usuario de ambos dataframes
for i in(0...df.size )
  
    f = df.row[i][1].split
    fecha = f[0]
  
    usuario = df.row[i][2]
  
    dato  =  [fecha, usuario]
  
    usuarios_y_fechas.push(dato)
end    


for i in(0...df2.size )
    f = df2.row[i][1].split
    fecha = f[0]
  
    usuario = df2.row[i][2]

    dato  =  [fecha, usuario]

    usuarios_y_fechas.push(dato)
end   


#se saca frecuencia de usuarios y fechas
unico = usuarios_y_fechas.uniq()

#hash_usuarios_y_fechas = usuarios_y_fechas.tally

#hash_usuarios_y_fechas = Hash[hash_usuarios_y_fechas.sort_by{|k, v| k[0]}]


hash_unico = unico.tally

hash_unico = Hash[hash_unico.sort_by{|k, v| k[0]}]

```

```ruby
nro_de_usuarios_por_fecha = Array.new


hash_unico.each do |key, value|
    nro_de_usuarios_por_fecha.concat( [key[0]] * value ) 
end

hash_nro_de_usuarios_por_fecha = nro_de_usuarios_por_fecha.tally

x = Array.new
y = Array.new

hash_nro_de_usuarios_por_fecha.each do |key, value|
    x.push(key) 
    y.push(value)
end


```
